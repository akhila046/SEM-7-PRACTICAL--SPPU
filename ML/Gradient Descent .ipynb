# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define Function and Derivative

```python
# Function
def f(x):
    return (x + 5)**2

# Derivative
def df(x):
    return 2 * (x + 5)
```

# Gradient Descent Implementation

```python
# Parameters
x = 3                 # Initial x
learning_rate = 0.01  # Step size
precision = 1e-6      # Convergence criteria
max_iterations = 10000

# Gradient Descent Loop
iteration = 0
x_history = [x]  # Store x values for reference

while iteration < max_iterations:
    gradient = df(x)
    x_new = x - learning_rate * gradient
    x_history.append(x_new)
    
    # Check convergence
    if abs(x_new - x) < precision:
        break
    
    x = x_new
    iteration += 1

print(f"Local minima occurs at x = {x_new}")
print(f"Minimum value of y = {f(x_new)}")
print(f"Total iterations: {iteration}")
```

# Observe Convergence

```python
for i, val in enumerate(x_history[:10]):  # Print first 10 iterations
    print(f"Iteration {i}: x = {val}, y = {f(val)}")
```

# Plot the Function and Gradient Descent Steps

```python

# Smooth function curve
x_vals = np.linspace(-10, 5, 200)
y_vals = f(x_vals)

# Plot function
plt.plot(x_vals, y_vals, label="y = (x+5)^2", color='blue')

# Plot gradient descent path (sample every 50 steps to reduce clutter)
step = max(1, len(x_history)//50)  # adjust step
plt.scatter(x_history[::step], [f(x) for x in x_history][::step], color='red', label="Gradient Descent Steps")

# Highlight start and end points
plt.scatter(x_history[0], f(x_history[0]), color='green', s=100, label="Start (x=3)")
plt.scatter(x_history[-1], f(x_history[-1]), color='orange', s=100, label=f"Minimum x={x_history[-1]:.2f}")

plt.plot(x_history[::step], [f(x) for x in x_history][::step], color='red', linestyle='--', alpha=0.5)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minima (Clean Visualization)")
plt.legend()
plt.grid(True)
plt.show()
```
