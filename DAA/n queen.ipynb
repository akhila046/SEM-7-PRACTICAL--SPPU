```python
def n_queens(n):
    # Sets to keep track of attacking columns and diagonals
    col = set()          # columns where queens are already placed
    posDiag = set()      # positive diagonals (r + c)
    negDiag = set()      # negative diagonals (r - c)

    res = []  # To store all solutions

    # Initialize board with "0" (no queens placed yet)
    board = [["0"] * n for i in range(n)]

    # Backtracking function
    def backtrack(r):
        # Base case: if we placed queens in all rows â†’ store solution
        if r == n:
            # Convert board rows into strings with spaces (for printing)
            copy = [" ".join(row) for row in board]
            res.append(copy)
            return

        # Try placing queen in every column of current row r
        for c in range(n):
            # Check if column or diagonals are under attack
            if c in col or (r + c) in posDiag or (r - c) in negDiag:
                continue  # skip this column

            # Place queen at (r, c)
            col.add(c)
            posDiag.add(r + c)
            negDiag.add(r - c)
            board[r][c] = "1"

            # Recurse to next row
            backtrack(r + 1)

            # Backtrack: remove queen and reset state
            col.remove(c)
            posDiag.remove(r + c)
            negDiag.remove(r - c)
            board[r][c] = "0"

    # Start backtracking from row 0
    backtrack(0)

    # Print all possible solutions
    for sol in res:
        for row in sol:
            print(row)
        print()


if __name__ == "__main__":
    n_queens(8)  # Solve for 8-Queens
```
